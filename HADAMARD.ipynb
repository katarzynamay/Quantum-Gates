# HADAMARD GATE

#### HADAMARD TRANSFORMATION

$H := \frac{\sigma_{x}+ \sigma_{z}}{\sqrt{2}}$      

$HH^{\dagger} =  H^{\dagger}H = I \quad$ (Unitarity)       
 
$H = H^{\dagger} \quad$ (Hermiticity)

##### **STANDARD BASIS $ \; \vert 0 \rangle ,\;  \vert 1 \rangle $**

$\vert\Psi\rangle = \alpha_{ 0} \vert0\rangle + \alpha_{ 1}  \vert1\rangle $  

$|\alpha_{0}|^2+|\alpha_{1}|^2 = 1 $  

$\vert 0 \rangle = $ 
$
\begin{pmatrix}
1 \\
0 \\
\end{pmatrix} \quad
$
$\vert 1 \rangle = $ 
$
\begin{pmatrix}
0 \\
1 \\
\end{pmatrix}
$  


$ \sigma_{x} = $
$
\begin{pmatrix}
0 & 1 \\
1 & 0 \\
\end{pmatrix} \quad
$
$ \sigma_{z} = $
$
\begin{pmatrix}
1 & 0 \\
0 & -1 \\
\end{pmatrix} 
$  

$\frac{\sigma_{x}+ \sigma_{z}}{\sqrt{2}} = $
$
\frac{1}{\sqrt{2}}
\begin{pmatrix}
0 & 1 \\
1 & 0 \\
\end{pmatrix} +
$
$
\frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 0 \\
0 & -1 \\
\end{pmatrix} =
$
$
\frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
1 & -1 \\
\end{pmatrix}
$   


**The matrix representation of the Hadamard gate in the standard basis**

$
H = \frac{1}{\sqrt{2}}\begin{pmatrix}
1 & 1 \\
1 & -1 \\
\end{pmatrix}
$  

##### APPLYING HADAMARD GATE

$  H \vert 0 \rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
1 & -1 \\
\end{pmatrix}
\begin{pmatrix}
1 \\
0 \\
\end{pmatrix} = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
1 \\
\end{pmatrix} = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
0 \\
\end{pmatrix} + \frac{1}{\sqrt{2}}
\begin{pmatrix}
0 \\
1 \\
\end{pmatrix} =
\frac{1}{\sqrt{2}} \vert0\rangle + \frac{1}{\sqrt{2}} \vert1\rangle =
\vert + \rangle
$  

$  H \vert 1 \rangle = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
1 & -1 \\
\end{pmatrix}
\begin{pmatrix}
0 \\
1 \\
\end{pmatrix} = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
-1\\
\end{pmatrix} = \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
0 \\
\end{pmatrix} - \frac{1}{\sqrt{2}}
\begin{pmatrix}
0 \\
1 \\
\end{pmatrix} =
\frac{1}{\sqrt{2}} \vert0\rangle - \frac{1}{\sqrt{2}} \vert1\rangle =
\vert - \rangle
$  

$ H \vert + \rangle = \vert 0 \rangle $   

$ H \vert - \rangle = \vert 1 \rangle $   

##### **HADAMARD BASIS $ \; \vert + \rangle ,\;  \vert - \rangle $**

$ \vert + \rangle= $
$
\frac{1}{\sqrt{2}}\begin{pmatrix}
1 \\
1 \\
\end{pmatrix} \quad
$  

$ \vert - \rangle =$
$
\frac{1}{\sqrt{2}}\begin{pmatrix}
 1 \\
-1 \\
\end{pmatrix}
$  

$ \vert 0 \rangle = 
\begin{pmatrix}
 0 \\
1 \\
\end{pmatrix} = 
\frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
1 \\
\end{pmatrix} + \frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
-1 \\
\end{pmatrix} =
\frac{1}{\sqrt{2}} \vert + \rangle + \frac{1}{\sqrt{2}} \vert - \rangle 
$  

$ \vert 1 \rangle = 
\begin{pmatrix}
1 \\
0 \\
\end{pmatrix} = 
\frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
1 \\
\end{pmatrix} - \frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}}
\begin{pmatrix}
1 \\
-1 \\
\end{pmatrix} =
\frac{1}{\sqrt{2}} \vert + \rangle - \frac{1}{\sqrt{2}} \vert - \rangle 
$



```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex, plot_bloch_multivector, plot_histogram
```

### Creating superposition of qubits using Hadamard Gate

### STATEVECTOR QISKIT SIMULATOR

```python
qr = QuantumRegister(1, "qbit")
qc = QuantumCircuit(qr)
state0 = Statevector(qc)

#Hadamard Gate
qc.h(qr)
state1 = Statevector(qc)

print("\n----------------------------- |0> --------------------------------\n")
display(state0.draw('latex'))
display(state0.draw('paulivec'))
#display(state0.draw('qsphere'))
display(state0.draw('bloch'))

print("\n----------------------------- H |0> ------------------------------\n")
display(state1.draw('latex'))
display(state1.draw('paulivec'))
#display(state1.draw('qsphere'))
display(state1.draw('bloch'))

print("\n-------------------------- CIRCUIT H |0> -------------------------\n")
display(qc.draw('latex', initial_state = True))
```

### IBM QPU / Simulators

```python
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2 as Sampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager 
service = QiskitRuntimeService()

qr = QuantumRegister(1, "qbit")
cr = ClassicalRegister(1, "cbit")
qc = QuantumCircuit(qr, cr)
state0 = Statevector(qc)

#Hadamard Gate
qc.h(qr)
state1 = Statevector(qc)

#qc.measure(qr, cr)
qc.measure_all()

backend = service.backend("ibmq_qasm_simulator")
print(
    f"Name: {backend.name}\n"
    f"Version: {backend.version}\n"
    f"No. of qubits: {backend.num_qubits}\n"
)
pm = generate_preset_pass_manager(backend=backend, optimization_level=0)
isa_circuit = pm.run(qc)
sampler = Sampler(backend)
job = sampler.run([isa_circuit])

print(f">>> Job ID: {job.job_id()}")
print(f">>> Job ID: {job.usage_estimation}")
print(f">>> Job Status: {job.status()}")
print(f">>> Job Result: {job.result()}")

print("\n---------------------------- |0> --------------------------------\n")
display(state0.draw('latex'))
display(array_to_latex(state0))

print("\n-------------------------- H |0> ---------------------------------\n")
display(state1.draw('latex'))
display(array_to_latex(state1))
```

```python
job = service.job(job.job_id())
job_result = job.result()

print(f">>> Job ID: {job.job_id()}")
print(f">>> Job Status: {job.status()}")

print("\n-------------------------- H |0> 1024 SHOTS -----------------------\n")


for idx, pub_result in enumerate(job_result):
    print(f"Result {idx}: {pub_result.data.meas.get_counts()}")
    display(plot_histogram(pub_result.data.meas.get_counts(), color='midnightblue', title="1024 SHOTS") )
```
